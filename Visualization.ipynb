```python
import pandas as pd
# Load dataset, assuming it's a tab-separated file despite the .xls extension
df = pd.read_csv('MARKETING_DATA.xls', sep='\t')
df
```

```python
df_drop=df.dropna()
df_drop
```

```python
c_n= df.columns
c_n
```

```python
# Fill missing numbers with the median value of each column--handle missing values
df = df.fillna(df.median(numeric_only=True))
```

```python
# Convert date column to datetime type---fix data type
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

```python
# Replace very high incomes with the column's median--remove outleyers
median_income = df['Income'].median()
df.loc[df['Income'] > df['Income'].quantile(0.99), 'Income'] = median_income
df
```

### **SQL Tasks**

```python
# 1 Query most responsive customer segments
```

```python
# Group by Education and Marital_Status, sum responses
responsive_segments = df.groupby(['Education', 'Marital_Status'])['Response'].sum()
# Sort descending and get top segments
top_segments = responsive_segments.sort_values(ascending=False).reset_index()
print(top_segments.head())
```

```python
# 2 Aggregate campaign response by marital status or education level
```

```python
marital_response = df.groupby('Marital_Status')['Response'].sum().reset_index()
print(marital_response)
```

```python
education_response = df.groupby('Education')['Response'].sum().reset_index()
print(education_response)
```

```python
# 3 Calculate average spending and conversion per segment
```

```python
# Define spending columns
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
              'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Calculate averages per segment
segment_analysis = df.groupby(['Education', 'Marital_Status']).agg(
    avg_spending=('Income', 'mean'),
    conversion_rate=('Response', 'mean'),
    **{f'avg_{col}': (col, 'mean') for col in spend_cols}
).reset_index()

print(segment_analysis)
```


### **Optional Advanced**
- Use clustering or t-tests (Python) to find statistically significant groups

```python
from sklearn.cluster import KMeans
from scipy.stats import ttest_ind

# Prepare features
features = ['Income', 'Recency', 'NumWebPurchases', 'NumStorePurchases']
X = df[features].fillna(df[features].mean())

# Cluster customers
kmeans = KMeans(n_clusters=2, random_state=0)
df['cluster'] = kmeans.fit_predict(X)

# Compare response rates between clusters
cluster_0 = df[df['cluster'] == 0]['Response']
cluster_1 = df[df['cluster'] == 1]['Response']

# Perform t-test
t_stat, p_value = ttest_ind(cluster_0, cluster_1)
print(f"T-test results: t-stat={t_stat:.4f}, p-value={p_value:.4f}")
```

- Create a simple conversion probability score using logistic regression or scoring logic

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Prepare features and target
X = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'Income', 'Recency']].fillna(0)
y = df['Response']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train logistic regression
model = LogisticRegression()
model.fit(X_scaled, y)

# Generate conversion scores
df['conversion_score'] = model.predict_proba(X_scaled)[:, 1]

# Show top segments by conversion probability
print(df[['Education', 'Marital_Status', 'conversion_score']].sort_values('conversion_score', ascending=False).head())

```

### Visualizations

1. Heatmaps of response rate by demographics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot table: average response rate by Education and Marital_Status
pivot = df.pivot_table(values='Response',
                      index='Education',
                      columns='Marital_Status',
                      aggfunc='mean')

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Response Rate by Education & Marital Status')
plt.show()
```

2. Bar charts: product category-wise total spending

```python
import matplotlib.pyplot as plt

# List of product columns
product_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
                'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Total spending per category
totals = df[product_cols].sum()

totals.plot(kind='bar', color='skyblue')
plt.ylabel('Total Spending')
plt.title('Total Spending by Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. Line plot of campaign performance over time

```python
import matplotlib.pyplot as plt

# Example funnel stages (replace with your actual logic)
total_customers = len(df)
responded = df['Response'].sum()
purchased = df[df['NumWebPurchases'] > 0].shape[0]

stages = ['Total Customers', 'Responded', 'Purchased']
counts = [total_customers, responded, purchased]

plt.figure(figsize=(6,4))
plt.bar(stages, counts, color=['#4F81BD', '#C0504D', '#9BBB59'])
plt.title('Simple Customer Funnel')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

4. Customer funnel visual or simple dashboard in Matplotlib/Plotly

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=('Total Spending', 'Response Rate by Education'))

# Bar: total spending
fig.add_trace(go.Bar(x=totals.index, y=totals.values, name='Total Spending'), row=1, col=1)

# Heatmap: response rate
fig.add_trace(
    go.Heatmap(z=pivot.values, x=pivot.columns, y=pivot.index, colorscale='YlGnBu', name='Response Rate'),
    row=1, col=2
)

fig.update_layout(title_text='Simple Customer Dashboard', height=500)
fig.show()
```
